```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams['figure.figsize']=[16,9]
exponent = 1
dir = "./FULL_normedcors_TEST_" + str(exponent)
```

```python
# Cor 0 exact
exact = dict(np.load(dir+"/EXACT_CORS.npz"))
E_AB_aah = exact['AB_aah_0']
E_BA_aah = exact['BA_aah_0']
E_AB_aha = exact['AB_aha_0']
E_BA_aha = exact['BA_aha_0']
ts = np.linspace(0,10**(-exponent)*(len(E_BA_aah)-1),len(E_BA_aah), endpoint=True)

print(np.mean(E_AB_aha))
print(np.std(E_AB_aha))


plt.clf()
plt.title("A")
# plt.ylim(0.99999,1.00005)
plt.plot(ts,E_AB_aha.real, label='aha R')
plt.plot(ts,E_AB_aah.real, label='aah R')
plt.plot(ts,E_AB_aha.imag, label='aha I')
plt.plot(ts,E_AB_aah.imag, label='aah I')
plt.legend()
plt.show()
plt.clf()
plt.title("B")
# plt.ylim(0.99995,1.00001)
plt.plot(ts,E_BA_aha.real, label='aha R')
plt.plot(ts,E_BA_aah.real, label='aah R')
plt.plot(ts,E_BA_aha.imag, label='aha I')
plt.plot(ts,E_BA_aah.imag, label='aah I')
plt.legend()
plt.show()

plt.clf()
plt.title("SUM")
#plt.ylim(-0.000000000000000000001,0.000000000000000000001)
plt.plot(ts,E_AB_aha.real/2+E_AB_aah.real/2, label='aah+aha A REAL')
plt.plot(ts,E_BA_aha.real/2+E_BA_aah.real/2, label='aah+aha B REAL')
plt.plot(ts,E_AB_aha.imag/2+E_AB_aah.imag/2, label='aah+aha A IMAG')
plt.plot(ts,E_BA_aha.imag/2+E_BA_aah.imag/2, label='aah+aha B IMAG')
plt.legend()
plt.show()
```
