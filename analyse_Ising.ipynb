```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
dir = './ISING/ISING_NEWNTU_sud_5_5_1.0_0.3044_0.01_200'
```

```python
XA,XB,ZA,ZB,ts=[],[],[],[],[]
for i in range(1000):
    try:
        ENV = dict(np.load(dir + ('/RHOA_{:05d}.npz').format(i)))
    except: continue
    ts.append(i*0.01)
    X = np.array([[0,1],[1,0]])
    Z = np.array([[1,0],[0,-1]])
    XA.append(np.trace(X @ ENV['rhoA']))
    XB.append(np.trace(X @ ENV['rhoB']))
    ZA.append(np.trace(Z @ ENV['rhoA']))
    ZB.append(np.trace(Z @ ENV['rhoB']))

plt.plot(ts,XA)
plt.plot(ts,XB)
plt.show()
plt.plot(ts,ZA)
plt.plot(ts,ZB)
plt.show()
```
