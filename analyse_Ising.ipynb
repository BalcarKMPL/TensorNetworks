```python
import numpy as np
from matplotlib import pyplot as plt
dir = './BH/BH/ISING_NTU_SERIAL_SVD_36_6_1.0_0.30444_0.005_400'

XA,XB,ZA,ZB,ts=[],[],[],[],[]
for i in range(1000):
    try:
        ENV = dict(np.load(dir + ('/RHOA_{:05d}.npz').format(i)))
    except: continue
    ts.append(i*0.01)
    X = np.array([[0,1],[1,0]])
    Z = np.array([[1,0],[0,-1]])
    XA.append(np.trace(X @ ENV['rhoA']))
    XB.append(np.trace(X @ ENV['rhoB']))
    ZA.append(np.trace(Z @ ENV['rhoA']))
    ZB.append(np.trace(Z @ ENV['rhoB']))

plt.plot(ts,XA)
plt.plot(ts,XB)
plt.show()
plt.plot(ts,ZA)
plt.plot(ts,ZB)
plt.show()
```
