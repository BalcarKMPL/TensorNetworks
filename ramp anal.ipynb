```python
import numpy as np
from matplotlib import pyplot as plt
from ncon import ncon
import colorsys
import CTMRG_better_4 as CTM
from numpy.linalg import norm, svd
plt.rcParams['figure.figsize'] = [20,12]
```

```python
k=30.0
fig, ax = plt.subplots()
for D in [4,6,8,10,12,14]:
    dir = './BH-cubicramp_3_' + str(D) + '_' + str(k) + '_0.1'
    ts, deltas = [], []
    try: SPECS = dict(np.load(dir + '/SPECS.npz'))
    except: continue
    for n in range(1000000000):
        try: PEPS = dict(np.load(dir + '/PEPS_{:05d}.npz'.format(n)))
        except: break
        tQ = 0.1 * 2**(k/10)
        ts.append(PEPS['t']/tQ)
        deltas.append(PEPS['NTUdelta']/0.1)
    print(D,"\t",n)
    ax.set_title('NTU deltas     tQ_exp = ' + str(k))
    ax.set_yscale('log')
    ax.scatter(ts,deltas,label="D = " + str(D),marker='.')
    ax.plot(ts,[1e-3]*len(ts),color='k',ls=':')
ax.plot([0,0],[1e-3,1e-5],color='k',ls=':')
ax.legend()
```
