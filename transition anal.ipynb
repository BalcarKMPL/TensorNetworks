```python
from matplotlib import pyplot as plt
import numpy as np
import CTMRG_better_4 as CTM

def file(D, k):
    return './BH-cubicramp_3_' + str(D) + '_' + str(float(k)) + '_0.1'
def findzero(x,y):
    _t = []
    x = np.array(x)
    y = np.array(y)
    for i in range(x.shape[0]):
        _t.append(x**i)
    _t = np.array(_t).T
    a = np.linalg.solve(_t,y)
    y0 = np.array([1]+[0]*(x.shape[0]-1)) @ a
    return y0
```

```python
for D in [4,6,8,10,12,14]:
    AA_trans, NN_trans, ks = [], [], []
    for k in np.arange(0,11,1):
        dirr = file(D, k)
        ts, AAs, NNs = [], [], []
        for i in range(0,1000):
            try:
                PEPS = dict(np.load(dirr + f'/PEPS_{i:05d}.npz'))
                print(PEPS['t'])
                ENV = dict(np.load(dirr + f'/RHOA_2_{i:05d}.npz'))
            except: continue
            ts.append(PEPS['t'])
            AAs.append(np.real_if_close(np.mean(ENV['vals2'][0]+ENV['vals2'][1])))
            NNs.append(np.real_if_close(np.mean(ENV['vals2'][2])))
        ks.append(k)
        if len(AAs) != 1:
            NN_trans.append(findzero(ts, NNs))
            AA_trans.append(findzero(ts, AAs))
        else:
            NN_trans.append(NNs[0])
            AA_trans.append(AAs[0])

    plt.plot(ks,AA_trans)
```
