```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams['figure.figsize']=[16,9]
exponent = 1
dir = "./LOCAL_LR_BH_PEPS_0.02_1.0_1.96"
arA1, arB1 = [], []

NN = np.diag([0,1,4])

for i in range(1000):
    try:
        env1 = dict(np.load(dir+("/RHOA_{:05d}.npz".format(i))))
    except:
        print(i)
        break
    arA1.append(np.trace(env1['rhoA'] @ NN)-1)
    arB1.append(np.trace(env1['rhoB'] @ NN)-1)

arA1 = np.array(arA1)
arB1 = np.array(arB1)
ts = np.linspace(0,0.005*(len(arA1)-2),len(arA1), endpoint=True)

plt.clf()
plt.title("Real")
plt.plot(ts,arA1.real, label='1 A Real')
plt.plot(ts,arB1.real, label='1 B Real')
plt.legend()
plt.show()

plt.clf()
plt.title("Imag")
plt.plot(ts,arA1.imag, label='1 A Real')
plt.plot(ts,arB1.imag, label='1 B Real')
plt.legend()
plt.show()
```
